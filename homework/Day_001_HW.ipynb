```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt


def mean_absolute_error(mae_yd, mae_yp):
    """
    計算 MAE
    Args:
        - mae_yd: 自建立實際值
        - mae_yp: 預測值
    Return:
        - mae: MAE
    """
    mae = MAE = sum(abs(mae_yd - mae_yp)) / len(mae_yd)
    return mae


def mean_squared_error(mse_yd, mse_yp):
    """
    計算 MAE(兩點間的距離)
    Args:
        - mse_yd: 自建立實際值
        - mse_yp: 預測值
    Return:
        - mse: MSE
    """
    # 將兩個陣列相減後, 取平方, 再將整個陣列加總成一個數字(sum), 最後除以y的長度(len), 因此稱為"均方誤差"
    mse = MSE = sum(pow(mse_yd - mse_yp, 2)) / len(mse_yd)
    return mse


def pow_test(val):
    return pow(val, 2)


if __name__ == '__main__':
    w = 3
    b = 0.5

    # 0-100分割101
    x_lin = np.linspace(0, 100, 101)

    # numpy.ndarray
    # print('[+] x_lin type: ', type(x_lin))

    # 常態分佈，取樣101個資料(一維)
    np_batch_rad = np.random.randn(101)

    # numpy.ndarray
    # print('[+] np_batch_rad type: ', type(np_batch_rad))
    # print('[+] np_batch_rad : ', np_batch_rad)

    # 假裝有一預測值np_batch_rad * 5
    y = (x_lin + np_batch_rad * 5) * w + b

    # 這邊就是將 x_lin 以及剛剛算完的 y, 當作座標值, 將101個點在平面上畫出來
    # b. : b 就是 blue, 點(.) 就是最小單位的形狀, 詳細可以查 matplotlib 的官方說明
    # plt.legend(loc=2) : loc = 2，產製的圖例位置 'upper left'   : 2
    plt.plot(x_lin, y, 'b.', label='data points')
    plt.title("Assume we have data points")
    plt.legend(loc=2)
    plt.show()

    # 假裝有一實際值
    y_hat = x_lin * w + b
    plt.plot(x_lin, y, 'b.', label='data')
    plt.plot(x_lin, y_hat, 'r-', label='prediction')
    plt.title("Assume we have data points (And the prediction)")
    plt.legend(loc=2)
    plt.show()

    MSE = mean_squared_error(y, y_hat)
    MAE = mean_absolute_error(y, y_hat)
    print("The Mean squared error is %.3f" % MSE)
    print("The Mean absolute error is %.3f" % MAE)
```
